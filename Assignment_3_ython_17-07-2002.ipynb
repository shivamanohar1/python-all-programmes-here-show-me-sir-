50.Write a Python program that uses multiple built-in functions together to process a list of numbers

```python
numbers = list(map(int, input("Enter numbers").split()))

total = sum(numbers)
largest = max(numbers)
smallest = min(numbers)
length = len(numbers)
sorted_numbers = sorted(numbers)
reversed_numbers = list(reversed(numbers))
print(total)
print(largest)
print(smallest)
print(len)
print(sorted_numbers)
print(reversed_numbers)
```
